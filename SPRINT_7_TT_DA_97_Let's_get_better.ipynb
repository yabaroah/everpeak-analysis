## 🚩 __CLASIFICACIÓN DE VALORES NULOS <code>NaN</code>__ 👽
---

* __MCAR (MISSING COMPLETLY AT RANDOM):__ Cada registro que es nulo lo hace sin nungún patron específico. Es decir, cualquiera de los registros tiene la misma chance de estar vacío. RECOMENDACIÓN: ELIMINAR O IMPUTAR (MEDIA O LA MEDIANA)

* __MAR (MISSING AT RANDOM):__ Hay valores nulos, pero estos se relacionan con otra columna específica. EJ: Hay valores nulos, pero estos se asocian a una o varias columnas.

* __MNAR (MISSING AT NO RANDOM):__ El valor no está o aparece nulo, pero no es al azar. Por ejemplo: Persona de áreas ricas de la ciudad, no dan su nombre por privacidad, por lo que la coluna user_name, suele tener nulos.
